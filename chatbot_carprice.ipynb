```python
def answer_question(message, history=None):
    msg = message.lower()

    if car_data.empty:
        return "Dataset is not loaded."

    # ---------- Most/Least driven Petrol or Diesel cars ----------
    if ("petrol" in msg or "diesel" in msg) and any(
        kw in msg for kw in (
            "most driven", "highest kms", "highest km", "max kms", "maximum kms",
            "least driven", "lowest kms", "minimum kms", "min kms",
            "least kilometers", "least kilometres", "low km", "min km")):

        fuel = "petrol" if "petrol" in msg else "diesel"
        subset = car_data[car_data["Fuel_Type"].str.lower() == fuel]

        if subset.empty:
            return f"No {fuel} cars in the dataset."

        wants_most = any(kw in msg for kw in (
            "most driven", "highest kms", "highest km", "max kms", "maximum kms"))

        kms_val = subset["Kms_Driven"].max() if wants_most else subset["Kms_Driven"].min()
        cars = subset.loc[subset["Kms_Driven"] == kms_val, "Car_Name"].unique()

        label = "Most" if wants_most else "Least"
        return f"{label}-driven {fuel} car(s) at {kms_val:,} km: " + ", ".join(cars)

    # ---------- Highest selling price ----------
    elif "highest" in msg or "most expensive" in msg or "costliest" in msg:
        top = car_data.loc[car_data['Selling_Price'].idxmax()]
        return f"The most expensive car is {top['Car_Name']} at ₹{top['Selling_Price']} lakhs."

    # ---------- Lowest selling price ----------
    elif "lowest" in msg or "cheapest" in msg or "least expensive" in msg:
        low = car_data.loc[car_data['Selling_Price'].idxmin()]
        return f"The cheapest car is {low['Car_Name']} at ₹{low['Selling_Price']} lakhs."

    # ---------- Selling price of a car ----------
    elif "selling price of" in msg:
        try:
            model = msg.split("selling price of")[1].strip()
            match = car_data[car_data['Car_Name'].str.lower() == model]
            if match.empty:
                return f"No selling price found for {model.title()}."
            return f"{model.title()} is listed for ₹{match['Selling_Price'].iloc[0]} lakhs."
        except:
            return "Please specify the car model (e.g. selling price of Swift)."

    # ---------- Kms driven ----------
    elif "kms" in msg or "kilometers" in msg:
        model = msg.split()[-1]
        match = car_data[car_data['Car_Name'].str.lower() == model]
        if match.empty:
            return f"No data for {model.title()}."
        return f"{model.title()} has been driven {match['Kms_Driven'].iloc[0]:,} kms."

    # ---------- Most-driven car(s) ----------
    elif any(kw in msg for kw in (
            "most driven", "highest kms", "highest km", "max kms", "maximum kms",
            "most kilometers", "most kilometres", "max km", "max kilometer")):
        max_km = car_data["Kms_Driven"].max()
        cars = car_data.loc[car_data["Kms_Driven"] == max_km, "Car_Name"].unique()
        return f"Most-driven car(s) at {max_km:,} km: " + ", ".join(cars)

    # ---------- Least-driven car(s) ----------
    elif any(kw in msg for kw in (
            "least driven", "lowest kms", "minimum kms", "min kms",
            "least kilometers", "least kilometres", "low km", "min km")):
        min_km = car_data["Kms_Driven"].min()
        cars = car_data.loc[car_data["Kms_Driven"] == min_km, "Car_Name"].unique()
        return f"Least-driven car(s) at {min_km:,} km: " + ", ".join(cars)

    # ---------- Fuel type of a car ----------
    elif "fuel type" in msg:
        model = msg.split("fuel type of")[-1].strip()
        match = car_data[car_data['Car_Name'].str.lower() == model]
        if match.empty:
            return f"No fuel type info for {model.title()}."
        return f"{model.title()} runs on {match['Fuel_Type'].iloc[0]}."

    # ---------- List all diesel cars ----------
    elif msg.strip() == "diesel":
        cars = car_data[car_data['Fuel_Type'].str.lower() == "diesel"]["Car_Name"].unique()
        return "Diesel cars: " + ", ".join(cars) if len(cars) else "No diesel cars found."

    # ---------- List all petrol cars ----------
    elif msg.strip() == "petrol":
        cars = car_data[car_data['Fuel_Type'].str.lower() == "petrol"]["Car_Name"].unique()
        return "Petrol cars: " + ", ".join(cars) if len(cars) else "No petrol cars found."

        # ---------- NEW ➎  Owner-category *and* Fuel-type listings ----------
    elif (("petrol" in msg) or ("diesel" in msg)) and any(
        kw in msg for kw in (
            "first hand", "first owner", "single owner", "0 owner",
            "second hand", "second owner", "2nd owner", "2 owner",
            "third hand",  "third owner",  "3rd owner", "3 owner",
            "fourth hand", "4th owner",    "4 owner"
        )):

        # --- identify fuel ---
        fuel = "petrol" if "petrol" in msg else "diesel"

        # --- identify owner slice ---
        if any(kw in msg for kw in ("first hand", "first owner", "single owner", "0 owner")):
            owner_filter = (car_data["Owner"] == 0)      # adjust to ==1 if your dataset codes first owner as 1
            owner_label  = "first-hand"
        elif any(kw in msg for kw in ("second hand", "second owner", "2nd owner", "2 owner")):
            owner_filter = (car_data["Owner"] == 1)
            owner_label  = "second-hand"
        else:   # third or more
            owner_filter = (car_data["Owner"] >= 2)
            owner_label  = "third-hand (or more)"

        subset = car_data[
            owner_filter &
            (car_data["Fuel_Type"].str.lower() == fuel)
        ]

        cars = subset["Car_Name"].unique()
        if len(cars):
            return f"{owner_label.capitalize()} {fuel} cars: " + ", ".join(cars)
        else:
            return f"No {owner_label} {fuel} cars found."


    # ---------- Manual ----------
    elif "manual" in msg:
        cars = car_data[car_data['Transmission'].str.lower() == "manual"]["Car_Name"].unique()
        return "Manual cars: " + ", ".join(cars) if len(cars) else "No manual cars found."

    # ---------- Automatic ----------
    elif "automatic" in msg:
        cars = car_data[car_data['Transmission'].str.lower() == "automatic"]["Car_Name"].unique()
        return "Automatic cars: " + ", ".join(cars) if len(cars) else "No automatic cars found."

    # ---------- Year ----------
    elif "year" in msg:
        for word in msg.split():
            if word.isdigit() and len(word) == 4:
                year = int(word)
                cars = car_data[car_data['Year'] == year]["Car_Name"].unique()
                return f"Cars from {year}: " + ", ".join(cars) if len(cars) > 0 else f"No cars from {year}."
        return "Please mention a valid year like 2014 or 2017."

    # ---------- After/Before year ----------
    elif "after" in msg or "before" in msg:
        words = msg.split()
        year = None
        for word in words:
            if word.isdigit() and len(word) == 4:
                year = int(word)
                break
        if year:
            if "after" in msg:
                cars = car_data[car_data['Year'] > year]["Car_Name"].unique()
                return f"Cars after {year}: " + ", ".join(cars) if len(cars) else f"No cars found after {year}."
            elif "before" in msg:
                cars = car_data[car_data['Year'] < year]["Car_Name"].unique()
                return f"Cars before {year}: " + ", ".join(cars) if len(cars) else f"No cars found before {year}."
        else:
            return "Please mention a valid year like 'after 2014' or 'before 2018'."

    # ---------- Seller type ----------
    elif "seller" in msg and "of" in msg:
        try:
            model = msg.split("of")[-1].strip()
            match = car_data[car_data['Car_Name'].str.lower() == model]
            if match.empty:
                return f"No seller info for {model.title()}."
            return f"{model.title()} is sold by a {match['Seller_Type'].iloc[0]}."
        except:
            return "Please specify car name (e.g. seller of Ciaz)."

    # ---------- Owner info ----------
    elif "owner" in msg:
        model = msg.split()[-1]
        match = car_data[car_data['Car_Name'].str.lower() == model]
        if match.empty:
            return f"No ownership info for {model.title()}."
        count = match['Owner'].iloc[0]
        return f"{model.title()} had {count} previous owner{'s' if count != 1 else ''}."

    # ---------- Present price more than ----------
    elif "present price more than" in msg:
        try:
            val = float(msg.split("more than")[-1].strip())
            cars = car_data[car_data['Present_Price'] > val]["Car_Name"].unique()
            return f"Cars with present price > ₹{val}: " + ", ".join(cars) if len(cars) else "No such cars found."
        except:
            return "Mention a valid price like 'present price more than 5'."

    # ---------- Present price less than ----------
    elif "present price less than" in msg:
        try:
            val = float(msg.split("less than")[-1].strip())
            cars = car_data[car_data['Present_Price'] < val]["Car_Name"].unique()
            return f"Cars with present price < ₹{val}: " + ", ".join(cars) if len(cars) else "No such cars found."
        except:
            return "Mention a valid price like 'present price less than 5'."

    # ---------- Fallback ----------
    return "Sorry, I didn’t understand that. Please rephrase or ask about price, kms, fuel, owner, etc."
```

```python
# Launch the chatbot interface using updated Gradio format
chatbot = gr.ChatInterface(fn=answer_question, title="Car ChatBot", chatbot=gr.Chatbot(type="messages"))
chatbot.launch(True)
```
